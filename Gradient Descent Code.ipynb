```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
print(X)
```

```python
Y = (X-5)**2 + 3
print(Y)
```

```python
# Visualize the funtion
plt.style.use("seaborn")
plt.plot(X, Y)
plt.show()
```

```python
# Gradient Descent
# We have to find at what value of X, Y will be minimum

# Step 1: Initialize x with any random value
x = 0

# Now we will find slope at that point if slope is negative we will increase x else we decrease
# until we get slope zero. Function is given below
# X = x - lr * df/dx
lr = 0.1
plt.plot(X, Y)
for i in range(30):
    grad = 2*(x-5) # df/dx
    x = x - lr * grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)
    
plt.show()

# As we can see once we approach to slope 0, our x is increasing very slowly
```

```python
# What if we start with positive slope
x = 9

lr = 0.1
plt.plot(X, Y)
for i in range(30):
    grad = 2*(x-5) # df/dx
    x = x - lr * grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)
    
plt.show()
```

```python
# What if we change the learning curve
x = 0
lr = 0.3
plt.plot(X, Y)
for i in range(30):
    grad = 2*(x-5) # df/dx
    x = x - lr * grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)
    
plt.show()

# As you can see larger learning rate will coverge x quicker
```

```python
# But there is a catch if learning rate is too large we may not converge we may actually diverge
x = 0
lr = 2
plt.plot(X, Y)
for i in range(30):
    grad = 2*(x-5) # df/dx
    x = x - lr * grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)
    
plt.show()
```
