```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# Generate sample data
def generateDataset(m):
    X = np.random.randn(m)*10 #Genrate number in the form of standard normal distribution
    noise  = np.random.randn(m)*10 #Noise we added to replicate real world data
    y = 3*X + 3 + noise #Let's assume our function is linear
    return X, y
```

```python
X, y = generateDataset(100)
print(X)
```

```python
def plotData(X, y, color="green", title="Data"):
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.scatter(X, y, c=color)
    plt.show()
```

```python
plotData(X, y)
```

Before we feed this data, data has to go through a process called **standardization** or **nomalization**.

We want the features to have 0 mean and 1 standard deviation.

##### Why we need this?
- The presence of feature value X in the formula will affect the step size of the gradient descent.
- Because sometimes we have different features which are based on different scales, example height and age.
- To ensure that the gradient descent moves smoothly towards the minima and that the steps for gradient descent are updated at the same rate for all the features, we scale the data before feeding it to the model.


Here's the formula for standardization
$$
X = \frac{X - X.mean()}{X.std()}
$$

$$
\text{i.e. } X = \frac{X - \mu}{\sigma}
$$


```python
def normaliseData(X):
    X = (X - X.mean())/X.std()
    return X
```

```python
plotData(X, y)
X = normaliseData(X)
plotData(X, y)
```

You can notice our scale has changed

Now we have to do train-test split.
If we have 100% of data we will partition in such a way some part of data is used for training and remaining part for testing.
Example we will use 80% data for training and 20% for testing.

```python
def trainTestSplit(X, y, split = 0.8):
    m = X.shape[0]   # number of samples we have
    data = np.zeros((m, 2))
    
    data[:,0] = X
    data[:,1] = y
    
    np.random.shuffle(data)
    
    split = int(m*split)
    
    #training data
    XT = data[:split,0]
    YT = data[:split,1]
    
    #testing data
    Xt = data[split: ,0]
    Yt = data[split: ,1]
    
    return XT, YT, Xt, Yt
```

```python
XT, YT, Xt, Yt = trainTestSplit(X, y)
```

```python
print(XT.shape, YT.shape)
print(Xt.shape, Yt.shape)
```

```python
plt.scatter(XT, YT, color="orange", label="Train data")
plt.scatter(Xt, Yt, color="blue", label="Test data")
plt.title("Train-Test Data")
plt.legend()
plt.show()
```

```python
def hypothesis(X, theta):
    return theta[0]+theta[1]*X


def error(X, y, theta):
    m = X.shape[0]
    e = 0
    for i in range(m):
        y_i = hypothesis(X[i], theta)
        e = e + (y[i] - y_i)**2
        
    return e/(2*m)


def gradient(X, y, theta):
    m = X.shape[0]
    grad = np.zeros((2,))
    
    for i in range(m):
        exp = hypothesis(X[i], theta) - y[i]
        grad[0] += exp
        grad[1] += exp*X[i]
        
    return grad/m


def train(X, y, learning_rate=0.08, maxItr=100):
    theta = np.zeros((2,))
    error_list = []
    for i in range(maxItr):
        grad = gradient(X, y, theta)
        error_list.append(error(X, y, theta))
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
    
    plt.xlabel("Iteration Number")
    plt.ylabel("Loss/Error")
    plt.plot(error_list)
    return theta
```

```python
theta = train(X, y)
```

Above graph shows how our error is decreasing

### R2 Score
- Coefficient of determination is a statistical measure of how well the regression predictions approximate the real data points.
- An R2 of 1 indicates that the regression predictions perfectly fit the data.

The idea of calculation R2 score is given below:
Worst line that we can predict is the line $y = y_{\text{mean}}$ that is the average value of the given samples.
Now we will calculate the distance of all the points from this line i.e.
$$ T_1 = \sum_{i=1}^{m} (y^{(i)} - y_{\text{mean}})^2$$

Let's say our predicted line after training the model is $y_p$, so we will calcuate the distance of all the points from this line i.e.
$$ T_2 = \sum_{i=1}^{m} (y^{(i)} - y_{\text{p}})^2 $$

Now our R2 score will be,
$$\text{R2 Score} = 1 - \frac{T_1}{T_2}$$

##### **IF R2 score is near to 1 then our predicted line is good and if it's close to zero then our predicted line is not so good**

```python
# Model evaluation
def r2Score(y, yp):
    ymean = y.mean()
    num = np.sum((y - yp)**2)
    denom = np.sum((y - ymean)**2)
    
    return (1-(num/denom))
```

```python
print(r2Score(Yt, yp))
```
